## LRU(LEAST RECENLY USED) CACHE IMPLEMENTATION WITH HASH(DICTIONARY) AND QUEUE 

```python
import time
import random
```

definition of a single data object that needs to be cached as class Page with appropriate entities

```python
class Page:
    def __init__(self, value, key):
        self.value = value
        self.key = key
        self.nex = None
        self.prev = None
```

definition of a lru cache with approapriate entities and functions to implement retrievel and insertion into the cache

```python
class LRUCache():
    
    def __init__(self, capacity):
        self.capacity = capacity
        self.hashtable = dict()
        self.Queue_front = None;
        self.Queue_back = None;
        self.i = 0;
    
    
    def get(self, key):  
        if(key in self.hashtable.keys()):
            node = self.hashtable[key]
            next_node = node.nex
            prev_node = node.prev
            if(next_node!=None and prev_node != None):   
                next_node.prev = prev_node
                prev_node.nex = next_node
            if(prev_node!=None and next_node == None): #makes a lot of difference to effiency by commenting
                return node
            if(prev_node==None and next_node!=None):
                self.Queue_front =  next_node;
                next_node.prev = None
            if(prev_node==None and next_node==None): #makes a lot of difference to efficiency by commenting
                return node
            
            node.prev = self.Queue_back
            node.nex = None;
            self.Queue_back = node
            
            return node  #I will check for None while using the class and doing stuff there
        else:
            return None
            
    def put(self, node):
        if(self.i >= self.capacity):  #remove from front if cache is full
            to_be_deleted = self.Queue_front
            if(to_be_deleted.nex != None):
                self.Queue_front = self.Queue_front.nex
                self.Queue_front.prev = None
            if(to_be_deleted.key in self.hashtable.keys()):
                self.hashtable.pop(to_be_deleted.key)
            self.i-=1
            
        if(self.i==0):
            self.hashtable[node.key] = node
            self.Queue_front = self.Queue_back = node;
            self.i+=1
        else:
            self.hashtable[node.key] = node
            node.prev = self.Queue_back
            self.Queue_back.nex = node
            self.Queue_back = node
            self.i+=1
            
    def cle(self):
        self.hashtable.clear()
        self.Queue_front = None
        self.Queue_back = None
        self.i = 0
        
    def prin(self):
        for i in self.hashtable.keys():
            print(self.hashtable[i].value)
```

Implementation and timing

```python
keys = [i for i in range(10000)]
```

```python
data_dump =list()
for i in keys:
    data_dump.append(Page(int(random.uniform(1000,2000)),i))
```

```python
data_dump
```

Sequence that will be used to test retrieval speed of the cache random.

```python
seq_for_retr = [int(random.gauss(10000/2,10)) for i in range(10000)]
```

Here gauss distribution of random is used to give more likeliness for some numbers to appear.This is so that we can give a fair assesment to cache technique

```python
seq_for_retr
```

```python
start_cache = time.time()
hits = 0
misses = 0
lru = LRUCache(30)
lru.cle()
for i in seq_for_retr:
    data = lru.get(i)
    
    if(data == None):
        print(data_dump[keys.index(i)].value)
        lru.put(data_dump[keys.index(i)])
        misses+=1
    else:
        print(data.value)
        hits+=1
    #lru.put(data_dump[keys.index(i)])
    #print(i)
end_cache = time.time()
time_cache = end_cache - start_cache
```

```python
print("hits :",hits)
print("misses :",misses)
```

```python
print("cached time :",time_cache)
```

```python
start_norm = time.time()

for i in seq_for_retr:
    print(data_dump[keys.index(i)].value)

end_norm = time.time()
time_norm = end_norm - start_norm
```

```python
print("time normal",time_norm)
```

```python
perc_increse_in_efficiency=(time_norm - time_cache)/time_norm*100
```
